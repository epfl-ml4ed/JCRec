```python
import pandas as pd
import json
```

```python


# Chemin vers votre fichier JSON
json_file_path = 'data/jobs.json'

# Fonction pour charger les données JSON et extraire les identifiants de job
def get_job_ids_from_json(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    # Extraire les clés du dictionnaire JSON, qui sont les identifiants de job
    job_ids = set(data.keys())
    
    return job_ids

# Utiliser la fonction et afficher les identifiants de job
job_ids = get_job_ids_from_json(json_file_path)
print(len(job_ids))
```

```python
import pandas as pd
import json

# Ouvrez le fichier JSON et chargez-le dans une variable
with open('data/resumes.json', 'r') as file:
    json_data = json.load(file)

# Transformez les données JSON en une liste de dictionnaires
data = []
for key, values in json_data.items():
    data.append({"id": key,
    "skills": values})


# Créez un DataFrame à partir de cette liste
df = pd.DataFrame(data)
# Affichez les premières lignes pour vérifier
display(df)
```

```python
import random

df['wanted_position'] = df.apply(lambda x: random.choice(list(job_ids)), axis=1)
df
```

```python
len(df.wanted_position.unique())
```

```python
import json

# Define the file path
file_path = "data/filename.json"

# Save the data as a JSON file
with open(file_path, "w") as file:
    json.dump(data, file)

# Print a message to confirm the file has been saved
print("Data saved as JSON file:", file_path)
```

```python
# Create a dictionary to store the data
json_data = {}

# Iterate over the DataFrame rows
for index, row in df.iterrows():
    # Create a list to store the skills
    skills = []
    for skill in row["skills"]:
        skills.append([skill[0], skill[1]])

    # Assign the wanted position directly to the variable
    wanted = row["wanted_position"]

    # Create a dictionary for each row
    data = {
        "skills": skills,
        "wanted": wanted
    }

    # Add the dictionary to the main dictionary with the id as the key
    json_data[row["id"]] = data

# Define the file path
file_path = "data/resumes_reward.json"

# Save the data as a JSON file
with open(file_path, "w") as file:
    json.dump(json_data, file)

# Print a message to confirm the file has been saved
print("Data saved as JSON file:", file_path)
```
